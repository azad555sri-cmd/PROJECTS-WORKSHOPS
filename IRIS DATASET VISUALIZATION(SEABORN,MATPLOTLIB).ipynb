# Combining Box and Strip Plots

```python
fig=plt.gcf()
fig.set_size_inches(8,5)
fig=sns.boxplot(x='Species',y='SepalLengthCm',data=ds)
fig=sns.stripplot(x='Species',y='SepalLengthCm',data=ds,jitter=True,edgecolor='gray')
```

```python
# Create the box plot
ax = sns.boxplot(x="Species", y="PetalLengthCm", data=ds)
# Add strip plot on top
ax = sns.stripplot(x="Species", y="PetalLengthCm", data=ds, jitter=True, edgecolor="gray")

# Check how many artists are available before accessing them
if len(ax.artists) >= 3:
    # Set color for the third box
    boxtwo = ax.artists[2]
    boxtwo.set_facecolor('yellow')
    boxtwo.set_edgecolor('black')
    
    # Set color for the second box
    boxthree = ax.artists[1]
    boxthree.set_facecolor('red')
    boxthree.set_edgecolor('black')
    
    # Set color for the first box
    boxone = ax.artists[0]  # Changed variable name to boxone for clarity
    boxone.set_facecolor('green')
    boxone.set_edgecolor('black')
else:
    print(f"Not enough box plot artists available. Only {len(ax.artists)} found.")

plt.show()
```

# KDE Plot

```python
# Create a kde plot of sepal_length versus sepal width for setosa species of flower.
sub = ds[ds['Species'] == 'Iris-setosa']

# Using the updated syntax for kdeplot
sns.kdeplot(
    x=sub['SepalLengthCm'],
    y=sub['SepalWidthCm'],
    fill=True,  # This replaces 'shade=True'
    thresh=0,   # This helps replace 'shade_lowest=False'
    cmap="plasma")

plt.title('Iris-setosa')
plt.xlabel('Sepal Length Cm')
plt.ylabel('Sepal Width Cm')

plt.show()
```

# Stacked Histogram

```python
ds['Species'] = ds['Species'].astype('category')
```

```python
ds.head()
```

```python
list1=list()
mylabels=list()
for gen in ds.Species.cat.categories:
    list1.append(ds[ds.Species==gen].SepalLengthCm)
    mylabels.append(gen)
    
h=plt.hist(list1,bins=30,stacked=True,rwidth=1,label=mylabels)
plt.legend()
plt.show()
```

# Area plot

```python
ds['SepalLengthCm'] = ds['SepalLengthCm'].astype('category')
```

```python
ds.head()
```

```python
# Convert categorical data back to numeric before plotting
# If ds['SepalLengthCm'] is categorical, convert it back to numeric
if pd.api.types.is_categorical_dtype(ds['SepalLengthCm']):
    ds['SepalLengthCm'] = ds['SepalLengthCm'].astype('float')

# Make sure all columns are numeric
ds['SepalWidthCm'] = ds['SepalWidthCm'].astype('float')
ds['PetalLengthCm'] = ds['PetalLengthCm'].astype('float')
ds['PetalWidthCm'] = ds['PetalWidthCm'].astype('float')

# Now plot the area chart
ds.plot.area(y=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'], alpha=0.4, figsize=(8, 5))

plt.show()
```
